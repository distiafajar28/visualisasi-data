**Import Library**

```python
import matplotlib.pyplot as plt
```

**DATASET**

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
product_A_sales = [10,20,15,25,30,45,40,50,60,55,65,70]
product_B_sales = [5,10,8,15,18,20,22,30,25,35,40,45]
```

**Single Line Plot**

```python
plt.plot(months, product_A_sales, label='Product A')
plt.title('Penjualan Product per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Multiple Line Plot & Customizations**

```python
plt.plot(months, product_A_sales, label='Product A', linestyle="--", marker='o', color='black')
plt.plot(months, product_B_sales, label='Product B', linestyle="-", marker='x', color='pink')

plt.title('Penjualan Product per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Jenis Garis untuk Menunjukkan Tren**

```python
product_C_sales = [18,2,25,28,32,38,42,45,48,52,56,60]
product_D_sales = [7,9,11,13,16,18,20,23,25,28,30,33]

plt.plot(months, product_A_sales, label='Product A', linestyle="--", color='black')
plt.plot(months, product_B_sales, label='Product B', linestyle="-", color='red')
plt.plot(months, product_C_sales, label='Product C', linestyle=":", color='blue')
plt.plot(months, product_D_sales, label='Product D', linestyle="-.", color='purple')

plt.title('Penjualan Product per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Subplo pada Line Chart**

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

axs[0].plot(months, product_C_sales, label='Product C', color='blue', marker='d')
axs[0].set_title('Penjualan Product C per Bulan')
axs[0].set_xlabel('Bulan')
axs[0].set_ylabel('Jumlah Penjualan')
axs[0].legend()
axs[0].grid(True)

axs[1].plot(months, product_C_sales, label='Product D', color='black', marker='s')
axs[1].set_title('Penjualan Product D per Bulan')
axs[1].set_xlabel('Bulan')
axs[1].set_ylabel('Jumlah Penjualan')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
def line_plot():
    fig, ax = plt.subplots()
    ax.plot(months, product_A_sales)
    ax.set_title('Penjualan Produk Per Bulan')
    ax.set_xlabel('Bulan')
    ax.set_ylabel('Jumlah Penjualan')
    plt.show()

line_plot()
```

```python
def customize_line_plot():
    fig, ax = plt.subplots()
    ax.plot(months, product_A_sales, label='Product A', color='red', linestyle='--', marker='o')
    ax.plot(months, product_B_sales, label='Product B', color='green', linestyle='-', marker='x')
    ax.set_title('Penjualan Produk Per Bulan')
    ax.set_xlabel('Bulan')
    ax.set_ylabel('Jumlah Penjualan')
    ax.legend()
    ax.grid(True)
    plt.show()

customize_line_plot()
```

```python
def trend_lines_plot():
    fig, ax = plt.subplots()
    ax.plot(months, product_A_sales, label='Product A Trend', linestyle='--', color='purple')
    ax.plot(months, product_B_sales, label='Product B Trend', linestyle='-', color='red')
    ax.set_title('Tren Penjualan Produk Per Bulan')
    ax.set_xlabel('Bulan')
    ax.set_ylabel('Jumlah Penjualan')
    ax.legend()
    ax.grid(True)
    plt.show()

trend_lines_plot()
```

```python
def subplots():
    fig, axs = plt.subplots(2, 1, figsize=(10, 8))

    # Plot for Product A
    axs[0].plot(months, product_A_sales, label='Product A', color='red', marker='o')
    axs[0].set_title('Penjualan Produk A Per Bulan')
    axs[0].set_xlabel('Bulan')
    axs[0].set_ylabel('Jumlah Penjualan')
    axs[0].legend()
    axs[0].grid(True)

    # Plot for Product B
    axs[1].plot(months, product_B_sales, label="Product B", color="red", marker="x")
    axs[1].set_title("Penjualan Produk B Per Bulan")
    axs[1].set_xlabel("Bulan")
    axs[1].set_ylabel("Jumlah Penjualan")
    axs[1].legend()
    axs[1].grid(True)

    plt.tight_layout()
    plt.show()

subplots()
```
